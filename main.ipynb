```python
import pandas as pd
```

```python
movies=pd.read_csv('/content/movies2024.csv')
```

```python
movies=movies[['MovieID', 'Title', 'Overview', 'Genres']]
```

```python
movies['Tags'] = movies['Overview']+movies['Genres']
```

```python
new_data  = movies.drop(columns=['Overview', 'Genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector=cv.fit_transform(new_data['Tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['Title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[1:6]:
    print(new_data.iloc[i[0]].Title)
```

```python
def recommend(movies):
    index = new_data[new_data['Title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    printed_titles = set()
    for i in distance[1:6]:
        movie_index = i[0]
        title = new_data.iloc[movie_index].Title
        if title != movies and title not in printed_titles:
            print(title)
            printed_titles.add(title)

```

```python
recommand("Iron Man")
```
